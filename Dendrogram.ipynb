```python
import pandas as pd

df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')

import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(15, 12))
dendrogram(Z)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import pandas as pd

df = pd.read_csv('../NIPS_1987-2015.csv')

Z = linkage(df.iloc[:, 1:], 'ward')

import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(15, 12))
dendrogram(Z)
plt.show()
```
